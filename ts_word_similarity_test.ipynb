```python
def make_prompt_dict(adata, prompt):
    '''
        Takes an adata with annotated adata.obs table, and a prompt, and generates a list of prompts to be used
        to either search databases or prompt a LLM
        
        Prompt should be formatted with a column name in brackets [], which will be the token that gets replaced
        with column names
        
        Returns a dictionary where keys are the original column anme and values are the newly formatted prompt
    '''
    prompt_dictionary = {}
    sub1 = "["
    sub2 = "]"

    # getting index of substrings
    idx1 = prompt.index(sub1)
    idx2 = prompt.index(sub2)

    column_name = ''
    # getting elements in between
    for idx in range(idx1 + len(sub1), idx2):
        column_name = column_name + prompt[idx]
 
    categories = list(adata.obs[column_name].cat.categories)
    for category in categories:
        prompt_dictionary[category] = prompt.replace("[" + column_name + "]", category)
    return prompt_dictionary
```

```python
def keywordify(string):
    nltk.download('stopwords')
    nltk.download('punkt')
    return_dict = {}
    r = Rake()
    my_text = string

    r.extract_keywords_from_text(my_text)
    keywordList           = []
    rankedList            = r.get_ranked_phrases_with_scores()
    for keyword in rankedList:
      keyword_updated = keyword[1].split()
      keyword_updated_string    = " ".join(keyword_updated[:2])
      keywordList.append(keyword_updated_string)
      if(len(keywordList)>9):
        break
    return(keywordList)
```

```python
def keywordify_dict(results_dict):
    return_dict = {}
    for key in results_dict.keys():
        if results_dict[key] == "No answer found for query":
            return_dict[key] = "NA"
        else:
            return_dict[key] = str(keywordify(results_dict[key]))
    return return_dict
```
